```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python

# Load the dataset
file_path = "/content/Ordinal_Encoded_Student_Data.csv"
df_selected = pd.read_csv(file_path)

# Clean and convert categorical columns to numerical values
# Ensure any unexpected string values are removed
df_selected = df_selected.replace({
    "Hours Studied:": np.nan, "Attendance:": np.nan, "Grade:": np.nan, "Previous Grade:": np.nan,
    "Grade Categories (Exam & Previous Scores):": np.nan
})

df_selected["Hours_Studied_Cat"] = df_selected["Hours_Studied_Cat"].replace({
    "3 - Long": 3, "2 - Average": 2, "1 - Short": 1
}).astype(float)

df_selected["Attendance_Cat"] = df_selected["Attendance_Cat"].replace({
    "3 - High (90-99)": 3, "2 - Average (80-89)": 2, "1 - Low (10-79)": 1
}).astype(float)

df_selected["Grade_Cat"] = df_selected["Grade_Cat"].replace({
    "1 - 95-99": 1, "2 - 90-94": 2, "3 - 85-89": 3, "4 - 80-84": 4,
    "5 - 75-79": 5, "6 - 70-74": 6, "7 - 65-69": 7, "8 - Below 65": 8
}).astype(float)

df_selected["Previous_Grade_Cat"] = df_selected["Previous_Grade_Cat"].astype(float)

df_selected["Motivation_Level"] = df_selected["Motivation_Level"].replace({
    "High - 3": 3, "Medium - 2": 2, "Low - 1": 1
}).astype(float)

# Ensure no string values exist in dataset
df_selected = df_selected.apply(pd.to_numeric, errors='coerce')

# Drop any remaining NaN values
df_selected.dropna(inplace=True)

# Convert cleaned data to integer type
df_selected = df_selected.astype(int)

# Display class distribution before training
print("\n--- Grade Distribution ---")
print(df_selected["Grade_Cat"].value_counts().rename_axis("Grade").reset_index(name="Count"))
print("\nGrade Mapping:")
print("1 - 95-99\n2 - 90-94\n3 - 85-89\n4 - 80-84\n5 - 75-79\n6 - 70-74\n7 - 65-69\n8 - Below 65")

# Feature Influence On Grades
feature_means = df_selected.groupby("Grade_Cat")[["Hours_Studied_Cat", "Attendance_Cat", "Motivation_Level", "Sleep_Hours", "Tutoring_Sessions"]].mean()
feature_means.index.name = "Grade"
feature_means.rename(columns={
    "Hours_Studied_Cat": "Avg. Hours Studied",
    "Attendance_Cat": "Avg. Attendance",
    "Motivation_Level": "Avg. Motivation Level",
    "Sleep_Hours": "Avg. Sleep Hours",
    "Tutoring_Sessions": "Avg. Tutoring Sessions"
}, inplace=True)
print("\n--- Feature Influence on Grades ---")
print(feature_means)
print("\nCategory Mappings:")
print("Hours Studied: 1 - Short, 2 - Average, 3 - Long")
print("Attendance: 1 - Low (10-79), 2 - Average (80-89), 3 - High (90-99)")
print("Motivation Level: 1 - Low, 2 - Medium, 3 - High")

# Select features for prediction
features = ["Hours_Studied_Cat", "Attendance_Cat", "Motivation_Level", "Sleep_Hours", "Tutoring_Sessions"]
X = df_selected[features]

# Define the target variable (Ordinal classification)
y = df_selected["Grade_Cat"]

# Split into training and test sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Naive Bayes model for classification
model = GaussianNB()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, zero_division=1)

# Print evaluation metrics
print("\n--- Model Evaluation ---")
print("Model Accuracy:", accuracy)
print("\nClassification Report:")
print(report)

```
